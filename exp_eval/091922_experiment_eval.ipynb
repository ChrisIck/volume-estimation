```python
import stat

sbatch_temp = '''#!/bin/bash

#SBATCH --job-name={experiment}
#SBATCH --nodes=1
#SBATCH --cpus-per-task=4
#SBATCH --mem=32GB
#SBATCH --time=47:59:59
#SBATCH --mail-user=chris.ick@nyu.edu
#SBATCH --export=NONE
#SBATCH --output="o_bs_09210-%j.out"
#SBATCH --gres=gpu:1

module purge
module load anaconda3/2020.07

source /home/ci411/.bashrc
conda activate s3d_env
python /home/ci411/volume_estimation/batched_jobs/{run}/re_eval/{experiment}.py
'''

pyfile_temp = '''from volume_estimation import evaluation

evaluation.evaluate_experiment('{run}', model_list=['{experiment}'])
'''

path_temp = '/home/ci411/volume_estimation/batched_jobs/{run}/re_eval/'

run = '092122_batch_size'
out_path = path_temp.format(run=run)

experiments = ['bs_09210', 'bs_09211', 'bs_09212', 'bs_09213', 'bs_09214', 'bs_09215']
runall = ''
for experiment in experiments:
    sbatch_file = sbatch_temp.format(experiment=experiment, run=run)
    py_file = pyfile_temp.format(experiment=experiment, run=run)
    
    with open(os.path.join(out_path, experiment+'.sbatch'),'w') as f:
        f.write(sbatch_file)
        
    with open(os.path.join(out_path, experiment+'.py'),'w') as f:
        f.write(py_file)
        
    runall += 'sbatch {}.sbatch\n'.format(experiment)

runall_path = os.path.join(out_path, 'runall.bat')
with open(runall_path, 'w') as f:
    f.write(runall)

st = os.stat(runall_path)
os.chmod(runall_path, st.st_mode | stat.S_IEXEC)
```
